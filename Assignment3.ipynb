```python
import pandas as pd
import numpy as np

df = pd.read_csv("/home/ubuntu/Desktop/TCOB42/DSBDAL/IRIS - IRIS.csv")
df
```

```python
species_list= ['Iris-setosa','Iris-virginica','Iris-versicolor']
```

```python
for species in species_list:
    print(f"\nStatistics for {species}:")
    species_data = df[df['species'] == species]
    print(species_data.describe())
    print("\nMedian\n",species_data.median())
    print("\nVariance\n",species_data.var())
    print("\nMode\n",species_data.mode())
    print("\n")
    
    
    
```
